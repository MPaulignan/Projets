```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Paramtres
sigma = 1.78  # cart-type estim
z = norm.ppf(0.975)  # pour un niveau de confiance de 95%

# Plage de tailles d'chantillons (nombre de retours)
n_values = np.arange(10, 1501, 100)

# Calcul de la marge d'erreur pour chaque taille d'chantillon
margin_of_error = z * sigma / np.sqrt(n_values)

# Calcul des bornes de l'intervalle de confiance autour d'une moyenne hypothtique (ex: moyenne = 3.17)
mean = 3.19
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

# Trac du graphique
plt.figure(figsize=(10, 6))
plt.plot(n_values, lower_bound, label='Borne infrieure (95%)', linestyle='--', color='blue')
plt.plot(n_values, upper_bound, label='Borne suprieure (95%)', linestyle='--', color='red')
plt.fill_between(n_values, lower_bound, upper_bound, color='gray', alpha=0.2)
plt.axhline(y=mean, color='black', linestyle='-', label='Moyenne estime (3.19)')

plt.title("volution de l'intervalle de confiance (95%) selon le nombre de retours")
plt.xlabel("Nombre de retours (taille de l'chantillon)")
plt.ylabel("Intervalle de confiance autour de la moyenne")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
